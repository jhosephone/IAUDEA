```python
# Paso 1: Importar las bibliotecas necesarias
import pandas as pd

# Paso 2: Cargar el archivo CSV desde Google Colab (debes subir el archivo en Colab)
from google.colab import files
uploaded = files.upload()

# Paso 3: Cargar el archivo CSV en un DataFrame
import io
df = pd.read_csv(io.BytesIO(uploaded['train.csv']))

# Paso 4: Mostrar las primeras filas del DataFrame para inspeccionar los datos
print("Primeras filas del DataFrame:")
df.head()

# Paso 5: Exploración básica de los datos

# Contar estudiantes por rendimiento global
performance_counts = df['RENDIMIENTO_GLOBAL'].value_counts()
print("\nConteo de estudiantes por rendimiento global:")
print(performance_counts)

# Contar estudiantes por departamento
department_counts = df['ESTU_PRGM_DEPARTAMENTO'].value_counts()
print("\nConteo de estudiantes por departamento:")
print(department_counts)

# Contar estudiantes por programa académico (mostrando solo los 5 más comunes)
program_counts = df['ESTU_PRGM_ACADEMICO'].value_counts().head()
print("\nConteo de estudiantes por programa académico (top 5):")
print(program_counts)
```
