```python
import pandas as pd
import numpy as np
import unicodedata
import re
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

def transformar_datos(dataset):
    """
    Limpia y preprocesa un conjunto de datos, transformando valores y manejando datos faltantes.
    """

    # Ajustar el formato de la columna PERIODO
    dataset['PERIODO'] = dataset['PERIODO'].astype(str).str[:4]

    # Función auxiliar para eliminar acentos y caracteres no deseados
    def limpiar_texto(cadena):
        if pd.notna(cadena):
            texto_sin_acento = unicodedata.normalize('NFKD', cadena).encode('ASCII', 'ignore').decode('utf-8')
            return re.sub(r'[¿?]', '', texto_sin_acento)
        return cadena

    # Agrupar valores según su frecuencia relativa
    def agrupar_por_frecuencia(tabla, columna, niveles=[0.05, 0.2]):
        proporciones = tabla[columna].value_counts(normalize=True)

        def asignar_categoria(valor):
            frecuencia = proporciones.get(valor, 0)
            for i, umbral in enumerate(niveles):
                if frecuencia < umbral:
                    return f'CATEGORIA_{i+1}'
            return 'CATEGORIA_ALTA'

        return tabla[columna].apply(asignar_categoria)

    # Normalizar y agrupar columnas específicas
    columnas_texto = ['ESTU_PRGM_ACADEMICO', 'ESTU_PRGM_DEPARTAMENTO']
    for columna in columnas_texto:
        dataset[columna] = dataset[columna].apply(limpiar_texto)
        dataset[columna] = agrupar_por_frecuencia(dataset, columna)

    # Convertir respuestas binarias a valores numéricos
    conversion_binaria = {'Si': 1, 'S': 1, 'No': 0, 'N': 0}
    columnas_binarias = ['FAMI_TIENEINTERNET', 'ESTU_PAGOMATRICULAPROPIO']
    for columna in columnas_binarias:
        dataset[columna] = dataset[columna].map(conversion_binaria).fillna(0)

    # Mapear niveles de estrato a valores numéricos
    estratos = {
        'Sin Estrato': 0, 'Estrato 1': 1, 'Estrato 2': 2,
        'Estrato 3': 3, 'Estrato 4': 4, 'Estrato 5': 5, 'Estrato 6': 6
    }
    dataset['FAMI_ESTRATOVIVIENDA'] = dataset['FAMI_ESTRATOVIVIENDA'].map(estratos)
    moda_estrato = dataset['FAMI_ESTRATOVIVIENDA'].mode()[0]
    dataset['FAMI_ESTRATOVIVIENDA'] = dataset['FAMI_ESTRATOVIVIENDA'].fillna(moda_estrato)

    # Convertir rangos a valores numéricos
    conversion_matricula = {
        'Menos de 500 mil': 0, 'Entre 500 mil y menos de 1 millón': 1,
        'Entre 1 millón y menos de 2.5 millones': 2, 'Entre 2.5 millones y menos de 4 millones': 3,
        'Entre 4 millones y menos de 5.5 millones': 4, 'Entre 5.5 millones y menos de 7 millones': 5,
        'Más de 7 millones': 6, 'No pagó matrícula': 7
    }
    conversion_horas_trabajo = {
        '0': 0, 'Menos de 10 horas': 1, 'Entre 11 y 20 horas': 2,
        'Entre 21 y 30 horas': 3, 'Más de 30 horas': 4
    }
    dataset['ESTU_VALORMATRICULAUNIVERSIDAD'] = dataset['ESTU_VALORMATRICULAUNIVERSIDAD'].map(conversion_matricula).fillna(0)
    dataset['ESTU_HORASSEMANATRABAJA'] = dataset['ESTU_HORASSEMANATRABAJA'].map(conversion_horas_trabajo).fillna(0)

    # Completar valores faltantes en la educación de padres con la moda
    moda_educ_padre = dataset['FAMI_EDUCACIONPADRE'].mode()[0]
    moda_educ_madre = dataset['FAMI_EDUCACIONMADRE'].mode()[0]
    dataset['FAMI_EDUCACIONPADRE'] = dataset['FAMI_EDUCACIONPADRE'].fillna(moda_educ_padre)
    dataset['FAMI_EDUCACIONMADRE'] = dataset['FAMI_EDUCACIONMADRE'].fillna(moda_educ_madre)

    # Generar variables dummy
    dataset = pd.get_dummies(dataset, columns=[
        'FAMI_EDUCACIONPADRE', 'FAMI_EDUCACIONMADRE',
        'ESTU_PRGM_ACADEMICO', 'ESTU_PRGM_DEPARTAMENTO'
    ], prefix=['ED_PADRE', 'ED_MADRE', 'PRG_ACADEMICO', 'PRG_DEPARTAMENTO'])

    # Convertir valores booleanos a enteros
    dataset = dataset.replace({True: 1, False: 0})

    return dataset
```
